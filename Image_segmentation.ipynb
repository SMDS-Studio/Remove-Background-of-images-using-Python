```python
from sklearn.datasets import load_sample_image
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
```

```python
china = load_sample_image("china.jpg")
data = china/255
height, width, channels = data.shape
reshaped_data = data.reshape((width*height, channels))

gmm = GaussianMixture(n_components = 2, random_state = 42)
gmm.fit(reshaped_data)

labels = gmm.predict(reshaped_data)
segmented_image = labels.reshape((height, width))

object_mask = (labels == 0).reshape((height, width, 1))
isolated_object = np.where(object_mask, data, 0)

plt.figure(figsize = (12, 4))
plt.subplot(1, 3, 1)
plt.imshow(data)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(segmented_image, cmap = "plasma")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(isolated_object)
plt.axis("off")

plt.show()
```
